**TRABALHANDO COM DATAS:**

```python
# Importando a biblioteca
import pandas as pd
```

```python
# Leitura dos arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
# Concatenar todos os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
#Transformar coluna em data do tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificar o tipo de dados de cada coluna
df.dtypes
```

```python
#Alterar o tipo de dados da coluna Data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
#Verificar o tipo de dados de cada coluna
df.dtypes
```

```python
# Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
# Agrupar por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# Criar nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
#Pegar amostra de 15 linhas
df.sample(15)
```

```python
# Extrair mês e dia
df["Mes_Venda"], df["Dia_Venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
#Pegar amostra de 15 linhas
df.sample(15)
```

```python
# Retornando a data mais antiga
df["Data"].min()
```

```python

# Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
#Pegar amostra de 15 linhas
df.sample(15)
```

```python
# Criando a coluna de trimestre
df["trimestre_vendas"] = df["Data"].dt.quarter
```

```python
#Pegar amostra de 15 linhas
df.sample(15)
```

```python
# Filtrar vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) &(df["Data"].dt.month == 3)]
```
